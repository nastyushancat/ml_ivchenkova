<h1>Задача 1. Дан набор из $p$ матриц размерностью $(n, n)$ и $p$ векторов размерностью $(n, 1)$, найти сумму произведений матриц на векторы. Написать тесты для кода

```python
import numpy as np
def sum_prod(X, V):
    sum_matrix = 0
    for i in range(len(X)):
        sum_matrix += np.dot(X[i], V[i])
    return sum_matrix

X1 = np.array([[1, 1], [1, 1]])
X2 = np.array([[1, 1], [0, 1]])
X3 = np.array([[1, 0], [0, 1]])
V1 = np.array([[1], [1]])
V2 = np.array([[1], [0]])
V3 = np.array([[1], [1]])

print(sum_prod([X1, X2, X3], [V1, V2, V3]))
print(sum_prod([], []))
```

<h1>Задача 2. Дана матрица M, напишите функцию, которая бинаризует матрицу по некоторому threshold (то есть, все значения большие threshold становятся равными 1, иначе 0). Напишите тесты для кода

```python
import numpy as np
def binarize(M, threshold):
    return np.where(M >= threshold, 1, 0)

print(binarize(np.array([[0.1, 0.6], [0.7, 0.3]]), 0.5))
print(binarize(np.array([[0.1, 0.6], [0.7, 0.3]]), 0.5))
print(binarize(np.array([[10, 10, -8], [17, -8, 10], [-11, 11, -11]]), -5))
```

<h1>Задача 3. Напишите функцию, которая возвращает уникальные элементы из каждой строки матрицы. Напишите такую же функцию, но для столбцов. Напишите тесты для кода

```python
import numpy as np
def unique_rows(matrix):
    return [list(np.unique(row)) for row in matrix]

def unique_columns(matrix):
    return [list(np.unique(column)) for column in np.transpose(matrix)]

print(unique_rows([1,1,1]))
print(unique_rows([[1, 2, 3], [4, 5, 6], [1, 2, 3]]))
print(unique_rows([[1, 2, 2], [5, 5, 6], [1, 2, 3]]))

print(unique_columns([[1, 2, 2], [5, 5, 6], [1, 2, 3]]))
print(unique_columns([[1, 2, 2, 5], [5, 5, 6, 8], [1, 2, 3, 6], [1, 4, 3, 9]]))
```

<h1>Задача 4. Напишите функцию, которая заполняет матрицу с размерами $(m, n)$ случайными числами, распределенными по нормальному закону. Затем считает мат. ожидание и дисперсию для каждого из столбцов и строк, а также строит для каждой строки и столбца гистограмму значений (использовать функцию hist из модуля matplotlib.plot)

```python
import numpy as np
import matplotlib.pyplot as plt

def matrix_normal():
    m = np.random.randint(2, 6)  
    n = np.random.randint(2, 6)  
    matrix = np.random.normal(size=(m, n)) 
    mean_rows = np.mean(matrix, axis=1)  
    mean_columns = np.mean(matrix, axis=0)  
    var_rows = np.var(matrix, axis=1)  
    var_columns = np.var(matrix, axis=0)  
    print("Мат. ожидание по строкам:", mean_rows)
    print("Мат. ожидание по столбцам:", mean_columns)
    print("Дисперсия по строкам:", var_rows)
    print("Дисперсия по столбцам:", var_columns)
    
    plt.figure(figsize=(12, 6))
    for i in range(m):
        plt.subplot(m, 2, 2*i+1)
        plt.hist(matrix[i], bins=20, color='green', edgecolor='black')
        plt.title(f'Row {i+1}')
    for j in range(n):
        plt.subplot(n, 2, 2*j+2)
        plt.hist(matrix[:, j], bins=20, color='blue', edgecolor='black')
        plt.title(f'Column {j+1}')
    
    plt.tight_layout()  
    plt.show() 

matrix_normal()
```

<h1>Задача 5. Напишите функцию, которая заполняет матрицу $(m, n)$ в шахматном порядке заданными числами $a$ и $b$. Напишите тесты для кода

```python
import numpy as np

def chess(m, n, a, b):
    matrix = np.full((m, n), a)
    matrix[::2, ::2] = b  
    matrix[1::2, 1::2] = b  
    return matrix
print(chess(3,3,4,5))
print(chess(10,10,1,0))
```

<h1>Задача 6. Напишите функцию, которая отрисовывает прямоугольник с заданными размерами (a, b) на изображении размера (m, n), цвет фона задайте в схеме RGB, как и цвет прямоугольника. Цвета также должны быть параметрами функции. Напишите аналогичную функцию но для овала с полуосями a и b. Напишите тесты для кода.
Примечание: уравнение эллипса (границы овала) можно записать как:
<h1>$\frac{(x-x_0)^2}{a^2}+\frac{(y-y_0)^2}{b^2}=1$

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_shape(shape_func, a, b, m, n, shape_color, background_color):
    image = np.full((m, n, 3), background_color, dtype=np.uint8)  
    shape = shape_func(a, b, m, n) 
    image[shape] = shape_color  
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def generate_rectangle(a, b, m, n):
    return np.s_[0:a, 0:b] 

def generate_ellipse(a, b, m, n):
    x0, y0 = m // 2, n // 2  
    mask = np.zeros((m, n), dtype=bool)
    for i in range(m):
        for j in range(n):
            if ((i - x0) / a)**2 + ((j - y0) / b)**2 <= 1:  
                mask[i, j] = True
    return mask

draw_shape(generate_rectangle, 100, 50, 200, 300, [255, 0, 0], [255, 255, 255])  
draw_shape(generate_ellipse, 100, 50, 200, 300, [0, 0, 255], [255, 255, 255])  
```
